<h1>Importacion de librerias necesarias</h1>

```python
#Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Cargamos el contenido del archivo
df = pd.read_csv('Clientes_Comportamiento.csv')
```

```python
#Seleccionamos las caracteristicas que vamos a utilizar para clustering
X = df[['Ingresos_Mensuales', 'Gasto_Anual']]
```

```python
#Impolementamos el algoritmo de KMeans con 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
#Agregamos los centroides a los clusters
centroids = kmeans.cluster_centers_
```

```python
#EVisualizamos los clusters
plt.figure(figsize=(8, 6))
for cluster in range(3):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['Ingresos_Mensuales'], cluster_data['Gasto_Anual'], label=f'Cluster {cluster + 1}')
```

```python
#Añadimos los centroides al grafico
plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='X', s=200, label='Centroides')
plt.title('Clusters de Clientes')
plt.xlabel('Ingresos Mensuales ($)')
plt.ylabel('Gasto Anual ($)')
plt.legend()
plt.grid()
plt.show()
```
